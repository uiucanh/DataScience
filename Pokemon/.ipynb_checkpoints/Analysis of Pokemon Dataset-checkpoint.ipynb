```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
# Distribution of types, generations
def draw_bars(category, figsize = (10,5), ylim = 200, palette = 'Set1', xlabel = "", ylabel = "", title = ""):
    plt.figure(figsize=figsize)
    ax = sns.countplot(category, data=df, palette = palette)
    for p in ax.patches:
        ax.annotate("%i" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=11, color='gray', xytext=(0, 20),
            textcoords='offset points')
    ax.set_ylim(0,ylim) #To make space for the annotations",
    plt.title(title, y=1.05)
    ax.set(ylabel = ylabel, xlabel=xlabel)
    sns.despine(left=True, top=False)

draw_bars('generation', xlabel='Generation', ylabel='Count', title='Distribution of Pokemon accross generation')
```

Gen 6 and 7 have the lowest number of new pokemons while gen 1 and gen 5 have the highest.

```python
fig, ax = plt.subplots(1,2, figsize=(14,5), sharey=True)
order = df['type1'].unique()
colors = ['#7AC74C', '#EE8130', '#6390F0', '#A6B91A', '#A8A77A', '#A33EA1', '#F7D02C', '#E2BF65',
         '#D685AD', '#C22E28', '#F95587', '#B6A136', '#735797', '#96D9D6', '#6F35FC', '#705746', '#B7B7CE', '#A98FF3']
plt.suptitle('Distribution of Primary Type and Secondary Type', fontsize=15)
sns.countplot(y='type1', data=df, ax=ax[0], order=order, palette = colors)
sns.countplot(y='type2', data=df, ax=ax[1], order=order, palette = colors)
ax[0].set_ylabel('Primary Type')
ax[1].set_ylabel('Secondary Type')
plt.subplots_adjust(wspace =0.3)
for a in ax:
    a.set_xlabel("Count")
    a.set_xlim(0,150)
    a.yaxis.labelpad=15
    a.xaxis.labelpad=15
    for p in a.patches:
        a.annotate("%i" % p.get_width(), (p.get_width(),p.get_y()),
                       ha='center', va='center', fontsize=11, color='gray', 
                       xytext=(15, -6.5), textcoords='offset points')
```

The most popular types for the primary type are Water and Normal. The least common primary type is Flying<br>
The most popular sub-type is Flying with 95 pokemons. <br>

```python
counts = dict(df['type1'].value_counts().add(df['type2'].value_counts(), fill_value=0))
plt.figure(figsize=(13,5))
ax = sns.barplot(x=list(counts.keys()), y=list(counts.values()), order=order, palette = colors)
for p in ax.patches:
    ax.annotate("%i" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center', fontsize=11, color='gray', xytext=(0, 10),
        textcoords='offset points')
ax.set_ylim(0,150) #To make space for the annotations"
plt.title("Distribution of both Primary and Sub types combined", fontsize=15)
plt.ylabel("Count", labelpad=15)
plt.xlabel("Type", labelpad=15)
```

The most popular type is Water and Normal the second. The least common types are Ice and Ghost.

```python
# Create dataframe of distribution of types within generation to plot
order = list(order)
gen_counts = {}
for i in range(1, 8):
    for t in order:
        gen_counts['Generation ' + str(i)] = [0] * 18
    counts = dict(df[df['generation'] == i]['type1'].value_counts().add(df[df['generation'] == i]['type2'].value_counts(), fill_value=0))
    for j in counts:
        gen_counts['Generation ' + str(i)][order.index(j)] = counts[j]
        
gen_df = pd.DataFrame.from_dict(gen_counts)
gen_df.index = order
gen_df = gen_df.transpose()
```

```python
sns.set_style('white')
sns.set_context("talk")
ax = gen_df.plot.barh(figsize=(12,15), width=0.9, color=colors, title='Distribution of types across generations')
ax.set_xlabel('Count')
```
